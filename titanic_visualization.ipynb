```python
# Titanic Data Visualization
import seaborn as sns
import matplotlib.pyplot as plt

# Carica dataset Titanic
titanic = sns.load_dataset('titanic')

# Prima visualizzazione: distribuzione per classe
sns.countplot(data=titanic, x='class')
plt.title('Passenger Distribution by Class')
plt.show()

# Seconda visualizzazione: classe vs sopravvivenza
sns.countplot(data=titanic, x='class', hue='survived')
plt.title('Survival Rate by Class')
plt.legend(title='Survived', labels=['No', 'Yes'])
plt.show()
```
